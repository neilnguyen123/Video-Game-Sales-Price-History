I tried using the Plain List API in r2 to pull all games; however, it requires a shop parameter. I need to get a list of all shops that the website aggregates for prices for the shop paremters. I will use the list of all shops to get list of all games listed.

Looking at the filter options on the website, there should be 62 unique shops.

Let's use the Covered Stores API to get a list of all shops.

```python
# Pulls all covered shops, using API
URL_3 = 'https://api.isthereanydeal.com/v01/web/stores/all/'
r3 = requests.get(URL_3)
r3_json = r3.json()
print(f"There are {len(r3_json['data'])} covered shops")
```

```python
stores_id = [r3['id'] for r3 in r3_json['data']]
stores_title = [r3['title'] for r3 in r3_json['data']]

# Initialize dataset with the created lists
data = {
    'stores_id': stores_id,
    'stores_title': stores_title
}

# Create DataFrame
df = pd.DataFrame(data)
df
```

```python
# Check for duplicates
print(f"There are {df.duplicated().sum()} duplicate rows.")
```

The API is unable to pull the remaining 17 shops (only 73% covered). This may be an issue but for now let's move forward with looping the 36 shops to get all the games.

```python
# Exporting store list as csv for later usage
df.to_csv('isthereanydeals_stores.csv', index=False)
```

```python
game_list = []
for shop in stores_id:
    URL_4 = f'https://api.isthereanydeal.com/v01/game/plain/list/?key={KEY}&shops={shop}'
    r4 = requests.get(URL_4)
    game_list.append(r4.json())
game_list
```

This is list of all games for the 36 shops. Games are listed by their plain titles (no spaces and all lower cases). Let's put this into a dataframe and remove duplicates.

```python
#stores_id = [game_list['id'] for r4 in game_list['data']]
game_list[1]['data'].values()
```

I'm checking out the JSON structure, so I can parse the game titles. Looks like I will have to run several iterations to parse the names.

```python
game_name_list = []

for game_json in game_list:
    for game_json_1 in game_json['data'].values():
        for game in game_json_1.values():
            game_name_list.append(game)
```

```python
print(f"There are {len(game_name_list)} game titles from the 45 shops.")
```

There may be some duplicate titles, so let's load all the titles into a dataframe and remove duplicates.

```python
data = {
    'game_name': game_name_list
}

df_game_names = pd.DataFrame(data)
df_game_names
```

```python
print(f"There are {df_game_names.duplicated().sum()} duplicate titles.")
```

```python
# Remove duplicate titles and return cleaned dataframe
df_game_names.drop_duplicates(inplace=True)
df_game_names
```
